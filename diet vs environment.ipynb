```python
import pandas as pd

# Load the original dataset
df = pd.read_csv("Results_21Mar2022.csv")
# Display the first few rows to understand the structure
df.head()
```

```python
# Define columns to drop:
# All columns starting with "sd_" (standard deviation values)
#'mc_run_id' and 'grouping', which are not needed for visualization
columns_to_drop = [col for col in df.columns if col.startswith("sd_")] + ['mc_run_id','grouping']
df_clean = df.drop(columns=columns_to_drop)
df_clean.head()
```

```python
# Rename columns to more intuitive and analysis-friendly names
df_clean = df_clean.rename(columns={
    "diet_group": "Diet",
    "sex": "Gender",
    "age_group": "Age_Group",
    "mean_ghgs": "GHG_Emissions",
    "mean_land": "Land_Use",
    "mean_watscar": "Water_Scarcity",
    "mean_eut": "Eutrophication",
    "mean_ghgs_ch4": "GHG_CH4",
    "mean_ghgs_n2o": "GHG_N2O",
    "mean_bio": "Biodiversity_Loss",
    "mean_watuse": "Water_Use",
    "mean_acid": "Acidification",
    "n_participants": "Participants"
})
df_clean.head()
```

```python
# Save the cleaned dataset to a new CSV file
df_clean.to_csv("Cleaned_Diet_Environment_Data.csv", index=False)
```
